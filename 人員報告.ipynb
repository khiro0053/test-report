```python
#カンパニーの読み込み

import pandas as pd
cmp_filename = "GLE_EIS34353.xlsx"
cmp_df = pd.read_excel(cmp_filename, sheetname=cmp_wb.sheetnames[0])
cmp_df
```

```python
cmp_df.columns
```

```python
cmp_df.columns = ['社員番号', '社員名称', '性別', '生年月日', '年齢', '採用発令日', '退職日付', '勤続年月(休職控除)','会社コード', '所属コード', '事業所名称', '役職名称', '予備０１名称', '職種名称', '休職区分']
cmp_df.head()

```

```python
def get_sliced_str(x):
    if x.会社コード ==  "GL000000":
        return x.社員番号[2:]
    else:
        return x.社員番号[3:]
        
```

```python
cmp_df['社員番号'] = cmp_df.apply(get_sliced_str, axis=1)
cmp_df
```

```python
cmp_id = cmp_df['社員番号'].values
cmp_id
```

```python
#クロノスの読み込み
import openpyxl
import pandas as pd
cronos_filename = "kinmu.xls"

crs_df = pd.read_excel(cronos_filename, sheetname = '月間出勤簿', header = None)
crs_df.head(10)

```

```python
crs_df = crs_df[9:] #9行目から必要だからいらない情報を削る
crs_df
```

```python
# データフレームのインデックスをユニークな値にする。
# ※基本的に、連結などをしていなければ元々ユニークな連番で
# 割り振られています。
crs_df.reset_index(drop=True, inplace=True)
```

```python
crs_df.head()
```

```python
crs_df.tail()#最終行を返す
```

```python
len(crs_df)
```

```python
crs_df_filtering = crs_df.dropna()#NaNを消す
crs_df_filtering
```

```python
crs_id = crs_df_filtering[0].values #ndarrayに変換
crs_id 
```

```python
# 退社リスト　差集合　第1引数に含まれていて第2引数に含まれない）
import numpy as np
retired_list = np.setdiff1d(cmp_id, crs_id) #カンパニーにはいるがクロノスにはいない。つまり退職者リスト
retired_list
```

```python
# 入社リスト　差集合（第1引数に含まれていて第2引数に含まれない）
import numpy as np
joined_list = np.setdiff1d(crs_id,cmp_id) #クロノスにはいるがカンパニーにはいない。勤務時間をしらべて働いていたらカンパニーに登録。
joined_list
```
